```python
from tkinter import *
pencere = Tk()  # Tkinter penceresini türkçedeki pencere kelimesine atamış olduk.
pencere.title("Bilimin Çakıl Taşları") # title metodu başlık ismi vermemizi sağlar.
pencere.geometry("400x400+500+150")   # pencerenin boyutlarını ve soldan ile üsten uzaklıklarını belirlemiş olduk.
pencere.resizable(True,True) # ilk True değei False olursa enine büyütme stop, ikincisinde boyuna stop olur.
pencere.minsize(200,200) # dışarıdan müdahalede, pencerenin minumum olabileği boyutu belirledik.
pencere.maxsize(600,600) # dışarıdan müdahalede, pencerenin maxsimum olabileği boyutu belirledik.

pencere.update() # Pencere boyutunu güncelleyerek ilk baştaki 200 piksel boyutlarına geri döndürür.

en = pencere.winfo_width()  # pencerenin enini piksel olarak gösterir.
boy = pencere.winfo_height() # pencerenin boyunu piksel olarak gösterir.
sol = pencere.winfo_x()     # pencerenin soldan olan uzaklığını piksel olarak gösterir.
ust = pencere.winfo_y()     # pencerenin üstten olan uzaklığını piksel olarak gösterir.

print("En : " + str(en))    # metin ile toplamak için numerik olan ifadeyi srting yaparak toplatma
print("Boy : " + str(boy))  # print komutu ilede penceremizde değil sonuçları programımızın sayfasından görüyoruz.
print("Sol : " + str(sol))
print("Üst : " + str(ust))

mainloop()
```

```python
import tkinter as tk      # Burada her zamanki gibi öncelikle gerekli modülü içe aktardık:

pencere = tk.Tk()         # Daha sonra Tk() sınıfı yardımıyla penceremizi oluşturduk:

def çıkış():              # ardından çıkış() adlı bir fonksiyon tanımladık:
    etiket['text'] = 'Bana müsade'     # Etiketin text parametresini ‘Merhaba gençler’ olarak değiştiriyoruz.
    düğme['text'] = 'Bekleyin pencere kapatılıyor...' #Düğmenin text parametresini ‘Bekleyin…’ olarak değiştiriyoruz.
    düğme['state'] = 'disabled'  #Düğmenin state parametresini disabled olarak değiştirerek düğmeyi basılamaz hale getiriyoruz.
    pencere.after(5000, pencere.destroy) #5 saniye sonra ise pencere.destroy komutunu ile pencerenin kapanmasını sağlıyoruz.

# çıkış() fonksiyonunu tanımladıktan sonra Label() 
etiket = tk.Label(text='Merhaba gençler')
etiket.pack()

# ve Button() düğmeleri aracılığıyla etiket ve düğmelerimizi oluşturuyoruz:

düğme = tk.Button(text='Çık', command=çıkış)
düğme.pack()

# Buna göre, düğmeye basıldığında, command parametresinin değeri olan çıkış() fonksiyonu çalışmaya başlayacak ve 
# fonksiyon gövdesinde tanımladığımız işlemler gerçekleşecek.

# Bildiğiniz gibi, bir program penceresinde, o programı kapatmayı sağlayacak düğmelerin yanı sıra, bir de en üst sağ 
# (veya sol) köşede program penceresini kapatan bir ‘X’ düğmesi bulunur. İşte bu ‘X’ düğmesine basıldığında da pencere
# kapanmadan önce çıkış() fonksiyonunun çalışması için şu kodu yazıyoruz:
# protocol() de tıpkı geometry() gibi, Tk() sınıfının metotlarından biridir. Bu metodu WM_DELETE_WINDOW argümanıyla birlikte 
# kullanarak, pencere üzerindeki ‘X’ düğmesine basıldığında neler olacağını tanımlayabiliyoruz.
pencere.protocol('WM_DELETE_WINDOW', çıkış)

# Son olarak da ana döngü mekanizmasını çalıştırıyoruz ve penceremizi görünür hale getiriyoruz:
pencere.mainloop()
```

```python
# Arayüzlerinizi olu¸stururken farklı imleç ¸sekilleri kullanmak da isteyebilirsiniz. 
# Bunun için kullanacağımız seçenek “cursor”. Mesela:

from tkinter import *
pencere = Tk()
etiket = Label(text="Deneme 1,2,3...", cursor="bottom_side")
etiket.pack()
mainloop()

# Burada, imleci pencere üzerine getirdi˘gimizde imlecin ters ok ¸sekli aldı˘gını görürüz.
# Kullanılabilecek imleç isimleri için ¸su sayfaya bakabilirsiniz: http://www.dil.univmrs.

# http://www.dil.univmrs.fr/~garreta/PythonBBSG/docs/Tkinter_ref.pdf# 
```

```python
from tkinter import *
class Pencere(tk.Tk):
    def __init__(self):
        super().__init__()
        self.protocol('WM_DELETE_WINDOW', self.çıkış)

        self.etiket = tk.Label(text='Merhaba gençler')
        self.etiket.pack()

        self.düğme = tk.Button(text='Çık', command=self.çıkış)
        self.düğme.pack()

    def çıkış(self):
        self.etiket['text'] = 'Elveda gençler...'
        self.düğme['text'] = 'Bekleyin...'
        self.düğme['state'] = 'disabled'
        self.after(2000, self.destroy)

pencere = Pencere()
pencere.mainloop()
```

```python
from tkinter import *
class Uygulama(object):
    def __init__(self):
        self.araclar()
    def araclar(self):
        self.etiket = Label(text="Dosyayı silmek istedi˘ginizden emin misiniz?")
        self.etiket.pack()
pencere = Tk()
uyg = Uygulama()
mainloop()
# Gördüğünüz gibi burada “etiket” öğesini ayrı bir fonksiyon olarak belirledik. Programınız
# büyüyüp birbirinden farklı öğeler içermeye başladığı zaman bu öğeleri yukarıdaki şekilde farklı
# bölümlere ayırmak, kodlarınızı daha düzenli bir hale getirecek, ileride kodlarınızın bakımını
# yapmayı kolaylaştıracaktır.
# Eğer bu sınıflı kodları anlamakta zorluk çekiyorsanız, geçen sene anlattığım
# “Nesne Tabanlı Programlama” konusunu gözden geçirebilirsiniz. Bu arada hatırlatayım, Tkinter’i
# kullanmak için sınıflı yapıları bilmek zorunda değilsiniz. Ben de zaten bu sayfalarda sınıflı
# ve sınıfsız kullanımları bir arada vereceğim. Eğer sınıflı yapıyı anlamazsanız, sadece sınıfsız
# örneklerle de ilerleyebilirsiniz. Ama yine de benim size tavsiyem, sınıfları öğrenmeniz yönünde
# olacaktır. Çünkü internet üzerinde belli bir seviyeden sonra bu sınıflı yapılar hep karşınıza çıkacaktır.
# Yani siz kullanmasanız da, sınıfları kullanmayı bilmeniz sizin için faydalı olacaktır. Bu
# arada dikkat ederseniz, yukarıda verdiğim sınıflı son iki kod, daha önce verdiğim sınıflı koddan
# daha yetkin. Dediğim gibi, yeni özellikler öğrendikçe, kodlarımızı adım adım standartlara
# daha yakın hale getireceğiz.
# Bu derste Tkinter’i kullanarak basit bir pencereyi nasıl oluşturabileceğimizi öğrendik. Örneklerimizi
# hem sınıflı hem de sınıfsız kodlarla gösterdik. Tabi ki sınıflı kodlar bu tür küçük örneklerde
# çok lüzumlu değildir. Ama büyük bir projede çalışırken sınıflı yapı size esneklik ve vakit
# kazandıracak, kodlarınızın bakımını daha kolay yapmanızı sağlayacaktır.
# Bu açıklamaları da yaptığıma göre başka bir bölüme vaktimiz varsa geçebiliriz. 
# Yada haftaya kaldığığımız yerden devam ederiz.
```
